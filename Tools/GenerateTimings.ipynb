```python
edited_timings  = dict()

for time in timings:  # reverse keys with values
    value = timings[time]
    edited_timings[value] = time

country_names = list(edited_timings.keys())
country_start = list()
country_end = list()

for i in range(len(country_names) - 1):  # set start, end and country name lists
    country_start.append(edited_timings[country_names[i]])
    country_end.append(edited_timings[country_names[i + 1]])
    
country_names.remove("The End")  # remove The End
```

```python
import pandas as pd
time_data = pd.DataFrame()  # Generate a dataframe based on this data
time_data['start'] = country_start
time_data['end'] = country_end
time_data['name'] = country_names
time_data['save_name'] = ["%03d.mp4" % x for x in range(len(time_data))]
```
